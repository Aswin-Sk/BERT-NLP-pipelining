```python
import pandas as pd
df=pd.read_csv('/content/train1.csv')
```

```python
for i,row in df.iterrows():
  df.context[i]=df.context[i][df.answer_start[i]-50:df.answer_end[i]+50]
```

```python

import json
for i, row in df.iterrows():
  d[i] = [{'context':str(df.loc[i]['context']), f"qas" : [{
    'id' : str(df.loc[i]['id']),
    'question' : str(df.loc[i]['question']),
    'is_impossible' : False,
    f"answers":[{'text':str(df.loc[i]['answer_text']),'answer_start':int(df.loc[i]['answer_start'])}],
  }],},]

```

```python
!pip install datasets
from datasets import Dataset
dataset=Dataset.from_pandas(df)
```

```python
result =json.dumps(d)
print(result)
```

```python
!pip install tensorboardX
import logging
from simpletransformers.question_answering import QuestionAnsweringModel, QuestionAnsweringArgs
```

```python
model_type="bert"
model_name="bert-base-cased"
```

```python
#model configuration
model_args=QuestionAnsweringArgs()
model_args.num_train_epochs=5
model_args.train_batch_size=16


```

```python
model=QuestionAnsweringModel(model_type,model_name,model_args)
```

```python
!pip install wandb
```

```python
model.train_model(dataset)
```

```python
a=pd.read_csv('/test1.csv')
a.to_json()
```

```python
results,text=model.eval_model(a)
```
